```python
def pokemon(client, number):
    client = pokepy.V2Client()
    j = client.get_pokemon(number)
    k = j[0]
    s = client.get_pokemon_species(number)[0]
    lang = [[t.name, t.language.name] for t in s.names]
    li = [k.id, k.name, k.height, k.weight, lang]

    li_types = [[x.slot, x.type.name] for x in j[0].types]
    li.append(li_types)

    return li
```

```python
pokemon(client, 2)
```

```python
from tqdm import tqdm
from time import sleep

px = []

for i in tqdm(range(1, 801)):
    sleep(3)
    px.append(pokemon(client, number=i))

px
```
